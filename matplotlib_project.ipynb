```python
df=pd.read_csv('data/gapminder.csv')
#df.head(10)
```

```python
life_exp = list(df['lifeExp'])
population = list(df['population'])
gdpPerCap = list(df['gdpPerCap'])
years = list(df['year'])
print(type(years))
plt.xlabel('years')

plt.plot(years , life_exp,'bs', label = 'life Expetancy')
plt.plot(years, population, 'g',label = 'Population')
plt.plot(years, gdpPerCap ,'ro', label = 'GDP per Cap')
plt.legend()
#plt.show()
plt.savefig('vis_shivani.png')
plt.savefig('vis_shivani.pdf')
#savefig('vis_shivani.png', bbox_inches='tight')
```
